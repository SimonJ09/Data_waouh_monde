```python
import pandas as pd
import os

# Définissez le chemin du dossier contenant les fichiers Excel
dossier = "Nouveaux codes postaux"
dataframes = []

for fichier in os.listdir(dossier):
    if fichier.endswith(".xlsx"):
        chemin_fichier = os.path.join(dossier, fichier)

        dataframe = pd.read_excel(chemin_fichier)
        dataframes.append(dataframe)

# Fusionnez les DataFrames en un seul DataFrame
merged_dataframe = pd.concat(dataframes, ignore_index=True)

# Supprimez les doublons en fonction de l'index
merged_dataframe = merged_dataframe[~merged_dataframe.index.duplicated(keep='first')]

# Affichez les noms des colonnes
print("Noms des colonnes :")
print(list(merged_dataframe.columns))

# Affichez les premières lignes du DataFrame fusionné
print("Aperçu du DataFrame fusionné :")
merged_dataframe
```

```python
import pandas as pd
import os

# Définissez le chemin du dossier contenant les fichiers Excel
dossier = "Nouveaux codes postaux"
dataframes = {}

# Listes pour stocker les noms des variables locales et des DataFrames
noms_variables_locales = []
noms_dataframes = []

for i, fichier in enumerate(os.listdir(dossier)):
    if fichier.endswith(".xlsx"):  
        nom_dataframe = os.path.splitext(fichier)[0].replace(" ", "_")  # Remplace les espaces par des soulignements
        chemin_fichier = os.path.join(dossier, fichier)
        
        # Chargez les données dans le DataFrame correspondant et renommez la variable locale
        nom_variable_dataframe = f"df_{nom_dataframe}"
        locals()[nom_variable_dataframe] = pd.read_excel(chemin_fichier)

        dataframes[nom_dataframe] = locals()[nom_variable_dataframe]
        
        noms_variables_locales.append(nom_variable_dataframe)
        noms_dataframes.append(nom_dataframe)

# Affichez les noms des DataFrames
print("Liste des noms de variables locales :")
print(noms_variables_locales)
```

```python
import re
import pandas as pd

def est_mobile(numero):
    motif_mobile = r'^(6\d{8}|\+336\d{8}|6\d{7}|\+336\d{7}|[1-5]\d{8]|[1-5]\.\d{2}\.\d{2}\.\d{2])'
    return bool(re.match(motif_mobile, str(numero)))  # Convertir en chaîne de caractères

def est_fixe(numero):
    # Motif pour les numéros de téléphone fixe avec 9 chiffres
    motif_fixe = r'^(0[1-5]\d{7}|0[8-9]\d{7}|[1-5]\d{7}|[6-9]\d{7])'
    return bool(re.match(motif_fixe, str(numero)))  # Convertir en chaîne de caractères

df_fixe = df[df['tel'].apply(est_fixe)].copy()

# Supprimer la colonne "N° Fixe" si elle existe déjà
if 'N° Fixe' in df_fixe.columns:
    df_fixe.drop(columns=['N° Fixe'], inplace=True)

# Renommer la colonne "N° Mobile" en "N° Fixe"
df_fixe.rename(columns={'tel': 'N° Fixe'}, inplace=True)
df_fixe.rename(columns={'N° Mobile': 'N° Mobile fiabilise'}, inplace=True)
df_fixe.rename(columns={'cp': 'Code postal'}, inplace=True)

# Créer un DataFrame pour les numéros de téléphone mobile
df_mobile = df[df['tel'].apply(est_mobile)].copy()
df_mobile.rename(columns={'N° Mobile': 'N° Mobile fiabilise'}, inplace=True)
df_mobile.rename(columns={'tel': 'N° Mobile'}, inplace=True)
df_mobile.rename(columns={'cp': 'Code postal'}, inplace=True)
```
